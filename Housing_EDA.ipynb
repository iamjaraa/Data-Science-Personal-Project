## Interpreting our Eploratory Data Analysis (EDA)

The housing dataset used in the Exploratory Data Analysis (EDA) consists of 13 columns, including key features such as the price and the number of bedrooms. The researcher’s initial step involved inspecting the dataset by obtaining both the first 5 rows and the last 5 rows to get a preliminary understanding of its structure.

### The shape function - 
Upon calling the shape method, it was revealed that the dataset consists of 545 rows and 14 columns. This provides a clear view of the dataset’s structure, with each row representing a unique property and each column representing specific attributes such as price, number of bedrooms, area, and parking facilities.

### The describe function -
To gain a deeper understanding of the numerical attributes within the dataset, the researcher utilized the describe function. This revealed some key insights about the properties, especially the price. For instance, the minimum price of a house in the dataset was found to be 1.75 million, corresponding to a modest 1-bedroom apartment with a single bathroom and no parking space. This suggests that, while the property is inexpensive, it may not be suitable for buyers seeking additional amenities, particularly parking.

In contrast, the most expensive house in the dataset was valued at 13.3 million. This property, a 6-bedroom house with 4 bathrooms and 4 stories, also includes a parking space capable of accommodating 3 cars. This stark difference in price highlights the significant variability in property values within the dataset. Additionally, the average cost of a house is approximately 4.7 million, providing a useful benchmark for understanding the general pricing trends in the data.

## Countplot --

The countplot function was employed to examine the distribution of properties with and without basements. The findings showed that more than half of the properties in the dataset do not have a basement, while a slightly smaller proportion of properties are basement-equipped. This indicates a potential gap in the housing market, where consumers may not have widespread access to basement-equipped homes, possibly due to regional preferences or construction trends.

Furthermore, the countplot revealed that most of the houses in the dataset are semi-furnished, with over 200 properties falling into this category. In contrast, furnished properties are less common, with just below 150 properties meeting this criterion. This could suggest a market preference for homes that require additional customization by the owner, rather than fully furnished properties.



## Heat Map

The heatmap analysis further reveals important correlations between various features of the dataset, particularly how they influence property prices. Price is most strongly influenced by area, followed by the number of bathrooms. Other features such as the number of stories, bedrooms, and parking spaces show moderate influence on the price.

Interestingly, the heatmap also confirms some expected correlations, such as between the number of bedrooms and bathrooms, and between area and parking. However, it is also evident that the number of stories has little to no direct correlation with the area of the property, suggesting that properties with more stories do not necessarily have larger areas. This insight could be valuable for developers and investors looking to optimize property features for pricing strategies.

The heatmap, therefore, provides valuable guidance for further analysis or model-building. For example, focusing on variables such as area and bathrooms, which show strong correlations with price, could improve the accuracy of predictions regarding property values.

### Box and Whiskers analysis
 
The box and whiskers plot of price reveals some interesting insights. The box is large and positioned towards the lower end of the price range, indicating that most properties are priced at the lower end of the spectrum. However, the significant spread of the box suggests considerable variation in property values. The median price is also closer to the lower end of the interquartile range (IQR), indicating that more than half of the properties are priced relatively low.

One of the most striking features of this analysis is the numerous outliers above the upper whiskers, extending to values around 12 million. These outliers represent high-end properties that significantly deviate from the general price range in the dataset, further emphasizing the diversity of property types and their corresponding prices.

In terms of other features such as area, bedrooms, bathrooms, stories, and parking, the box and whiskers plots reveal that these variables are much smaller in scale compared to price, and are closely clustered near zero. While there are a few outliers indicating properties with unusually high numbers of these features, they are still relatively modest compared to the range of property prices.
  

## Scatter plot Analysis

The scatter plot analysis of area versus price reveals a positive correlation, with larger properties tending to have higher prices. The data points primarily cluster in the lower-left region, with area values between 2000 and 8000 and price values generally below 6 million. This suggests that the majority of properties in the dataset are of moderate size and relatively low price.

However, a few data points deviate from this cluster, particularly in the higher price and area ranges. For example, some properties with an area exceeding 10,000 square feet have prices above 10 million. These properties represent outliers in the dataset, likely luxury homes that skew the overall trend.

As area increases, the price variability also widens. Properties with larger areas exhibit greater price variability, suggesting that additional factors such as location, amenities, or property condition may also influence the price of larger homes. This observation could help investors and developers identify potential price-driving features in larger properties.
